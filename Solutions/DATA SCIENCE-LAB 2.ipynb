```python
'''
Question 1 :
Does neighbourhood (“which represents location”) affect the price of property or not? If yes how?
-From the question, neighbourhood and price columns is what will be used to find a solution
'''
```

```python
#Checking the total number of neighbourhoods in the dataset
df['neighbourhood'].nunique()
```

```python
#Displaying the first 10 rows of the two columns
df[['neighbourhood', 'price']].head(10)
```

```python
#Grouping by price and neighbourhood and also trying to find the average price for each neighbourhood 
question1 = df.groupby(['neighbourhood'])['price'].mean()
```

```python
#Reseting index and sorting the values from the highest price
question1 = question1.reset_index().sort_values(by='price', ascending=False)
```

```python
#Renaming price column to average_price since contents of that column are average prices from each neighbourhood
question1 = question1.rename(columns={'price':'average_price'} )
```

```python
#Showing the first 5 rows to ensure the new dataframe is as expected
question1.head()
```

```python
#Checking the number of rows to ensure it matches the total number of neighbourhoods in the dataset
question1.shape
```

```python
#Dividing the dataframe into 4 equal samples to visualize the data
sample1 = question1.sample(frac=0.25, random_state=0)
sample2 = question1.sample(frac=0.25, random_state=1)
sample3 = question1.sample(frac=0.25, random_state=2)
sample4 = question1.sample(frac=0.25, random_state=3)
```

```python
#Visualizing the first sample
sample1.plot.line(title='Average price against Neighbourhood Sample 1',
               x='neighbourhood', y='average_price',
               ylabel='price', xlabel='neighbourhood',figsize=(15, 8))
```

```python
#visualizing the second sample
sample2.plot.line(title='Average price against Neighbourhood Sample 2',
               x='neighbourhood', y='average_price',
               ylabel='price', xlabel='neighbourhood', figsize=(15, 8))
```

```python
#Visualizing the third sample
sample3.plot.line(title='Average price against Neighbourhood Sample 3',
               x='neighbourhood', y='average_price',
               ylabel='price', xlabel='neighbourhood', figsize=(15, 8))
```

```python
#Visualizing the fourth sample
sample4.plot.line(title='Average price against Neighbourhood Sample 4',
               x='neighbourhood', y='average_price',
               ylabel='price', xlabel='neighbourhood', figsize=(15, 8))
```

```python
'''
Question 4 :
-What neighbourhood should we focus on to increase our sales? Explain your answer.
'''
```

```python

#Calculating the summary statistics for each neighbourhood
#Using groupby() function to group the columns in accordance with each neighbourhood(218 of them)
neighbourhood_stats = df.groupby('neighbourhood')['price'].agg(['mean', 'count']).reset_index()

display(neighbourhood_stats)
```

```python
#Renaming mean and count columns to avg_price and num_properties respectively
neighbourhood_stats = neighbourhood_stats.rename(columns={'mean': 'avg_price','count': 'num_properties'})
display(neighbourhood_stats)
```

```python
#Checking for any null values that may lead to innacurate results
neighbourhood_stats.isnull().any()
```

```python
#Ranking neighbourhoods by average price and number of properties sold
#Assigning the results in price_rank and properties_rank columns respectively
neighbourhood_stats['price_rank'] = neighbourhood_stats['avg_price'].rank(ascending=False)
neighbourhood_stats['properties_rank'] = neighbourhood_stats['num_properties'].rank(ascending=False)

#Calculating the mean of price_rank and properties_rank for each row
#The values are then assigned to a new column called combined_rank
# Combined rank for the neighbourhoods arranged from lowest
neighbourhood_stats['combined_rank'] = neighbourhood_stats[['price_rank', 'properties_rank']].mean(axis=1)
neighbourhood_stats = neighbourhood_stats.reset_index().sort_values('combined_rank', ascending=True)

```

```python
# Top recommended neighbourhoods
top_neighbourhoods = neighbourhood_stats.head(10)
display(top_neighbourhoods)
'''
-These neighbourhoods are top recommended since from the data, it shows the demand is high hence high number of properties
 which leads to increase in prices of properties hence a high average price
-Combined rank helps find neighbourhoods with a balance of both to identify high yield income areas
'''
```

```python
# Least recommended neighbourhoods 
bottom_neighbourhoods = neighbourhood_stats.tail(10)
display(bottom_neighbourhoods)
'''
-These neighbourhoods are least recommended since from the data, it shows the demand is low hence low number of properties
 which leads to low prices of properties hence a low average price
-Combined rank helps find neighbourhoods with a balance of both to identify low yield income areas
'''
```
